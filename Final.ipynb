Importing Libraries

```python
import pandas as panda
import numpy as np  
import matplotlib.pyplot as plot
from sklearn import linear_model
from sklearn import metrics  
```

Importing and Reading the CSV File

```python
url = "http://bit.ly/w-data"
data = panda.read_csv(url)
```

```python
data
```

Visualizing the Data

```python
plot.scatter(data.Hours,data.Scores)
plot.xlabel('HOURS',fontsize=14,color='black')  
plot.ylabel('SCORED',fontsize=14,color='black')  
plot.show()
```

Training the Algorithm

```python
 h = data.drop('Scores',axis='columns')
```

```python
t = linear_model.LinearRegression()
```

```python
t.fit(h,data.Scores)
```

```python
t.coef_
```

```python
t.intercept_
```

```python
plot.scatter(data.Hours,data.Scores)
plot.xlabel('HOURS',fontsize=14,color='black')  
plot.ylabel('SCORED',fontsize=14,color='black')
plot.plot(data.Hours,t.predict(h),color='black')
plot.show()
```

Actual Score and Predicted Score

```python
data['Predicted Scores']=t.predict(h)
```

```python
data
```

```python
hours = 9.25
t.predict([[hours]])
```

```python
plot.scatter(data.Hours,data.Scores)
plot.xlabel('HOURS',color='black',fontsize=14)  
plot.ylabel('SCORED',color='black',fontsize=14)
plot.plot(data.Hours,t.predict(h),color='green')
plot.scatter(hours,t.predict([[hours]]),color='black')
```

```python
print('Mean Absolute Error :',metrics.mean_absolute_error(data.Scores,t.predict(h))) 
print('Mean Squared Error :',metrics.mean_squared_error(data.Scores,t.predict(h)))
```
